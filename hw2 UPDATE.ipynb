```python
##Q1
def tuple01(n = int, k = int, low = int, high = int):
    list01 = np.random.randint(low, high, size=(n,k), dtype=int)
    tuple01 = [tuple(i) for i in list01]
    return tuple01
    
tuple01(4,5,0,10)
```

```python
##Q2
##A
def question2_a(sample_list, x, y):
    op = []
    for m in range(len(sample_list)):
        li = [sample_list[m]]
        for n in range(len(sample_list)):
            if (sample_list[m][x] == sample_list[n][x] and sample_list[m][y] != sample_list[n][y]):
                li.append(sample_list[n])
        op.append(sorted(li, key=lambda dd: dd[y], reverse=True)[0])
    res = list(set(op))
    return res

question2_a([(1, 3, 5), (0, 1, 2), (1, 9, 8)], 0, 2)
```

```python
##B
def question2_b(sample_list, x, y):
    op = []
    for m in range(len(sample_list)):
        li = [sample_list[m]]
        for n in range(len(sample_list)):
            if (sample_list[m][x] == sample_list[n][x] and sample_list[m][y] != sample_list[n][y]):
                li.append(sample_list[n])
        op.append(sorted(li, key=lambda dd: dd[y], reverse=True)[0])
    return list(set(op))

question2_a([(1, 3, 5), (0, 1, 2), (1, 9, 8)], 0, 2)
```

```python
##C
def question2_c(sample_list, a, b):
    d01 = defaultdict(list)
    for i in sample_list:
        d01[i[a]].append(i)
    res = [max(val, key=lambda x: x[b]) for val in d01.values()]
    return res

question2_c([(1, 3, 5), (0, 1, 2), (1, 9, 8)], 0, 2)
```

```python
##D
start1 = time.time() 
for i in range(1000):
    question2_a(tuple01(4,3,0,10), 0, 2) 
end1 = time.time()
time1 = end1- start1
print(time1)

start2 = time.time() 
for i in range(1000):
    question2_b(tuple01(4,3,0,10), 0, 2) 
end2 = time.time()
time2 = end2- start2
print(time2)

start3 = time.time() 
for i in range(1000):
    question2_c(tuple01(4,3,0,10), 0, 2) 
end3 = time.time()
time3 = end3- start3
print(time3)

print('In the study, each function run 1000 times with random tuple lists, and method 2a use' + str(time1) + ' minutes,and in 2b use' + str(time2) + ' minutes,and in 2c use' + str(time3) + ' minutes')
```
